```python
mhws = pd.read_csv(data_path + 'event_lon_{}_lat_{}.csv'.format(lon, lat), parse_dates=[3, 4, 5, 6])
mhws['row'] = range(len(mhws))
starts = mhws[['date_start', 'row']].rename(columns={'date_start': 'date'})
ends = mhws[['date_end', 'row']].rename(columns={'date_end':'date'})
mhws_decomp = pd.concat([starts, ends])
mhws_decomp = mhws_decomp.set_index('row', append=True)
# mhws_decomp.sort_index()[0:10]
```

```python
mhws_decomp = mhws_decomp.groupby(level=[0, 1]).apply(lambda x: x.set_index('date').resample('D').fillna(method='pad'))
mhws_decomp = mhws_decomp.reset_index(level=1, drop=True).reset_index()
mhws_decomp
```

```python
mhws_days = mhws_decomp.groupby(mhws_decomp['date'].dt.to_period('Y'))
mhws_days = mhws_days.nunique().date[-15:-1]
print(mhws_days)
print(np.array(mhws_days))
mhw_days = np.array([ 15, np.nan, 9,  9, 20,  21,  6,  np.nan, np.nan, np.nan,  5,  24,  26, 180, 229,  54,   9, 115])
len(mhw_days)
```
